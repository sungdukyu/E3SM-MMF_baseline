```python
import numpy as np
import csv
import matplotlib.pyplot as plt
import glob
import pandas as pd
from tensorflow import keras
```

```python
MODELPATH=pd.read_csv('../step2_retrain/top_models_after_step2.csv', index_col=0, header=None)
```

```python
for k in range(1,17):
    f_model = f'../step2_retrain/{MODELPATH.loc[k].values.item()}'.replace(' ','')
    model = keras.models.load_model(f_model, compile=False)
    model.summary()
    
    f_input = '/pscratch/sd/s/sungduk/val_input_stride6.npy'
    ml_in = np.load(f_input)
    
    ml_out = model.predict(ml_in)
    
    f_save = f'./predicted/{k:03}_{f_model.split("/")[-2]}_{f_model.split("/")[-1]}.npy'
    np.save(f_save, ml_out)
    
```
